```python
import pandas as pd
```

```python
x_b3lyp = pd.read_csv('../dataSets/X_b3lyp.csv', header=None)
y_b3lyp = pd.read_csv('../dataSets/Y_b3lyp.csv', header=None)
q_b3lyp = pd.read_csv('../dataSets/Q_b3lyp.csv', header=None)
```

```python
x_cc = pd.read_csv('../dataSets/X_cc.csv', header=None)
y_cc = pd.read_csv('../dataSets/Y_cc.csv', header=None)
q_cc = pd.read_csv('../dataSets/Q_cc.csv', header=None)
```

```python
x_b3lyp.head()
```

```python
x_cc.head()
```

```python
y_b3lyp.columns=['energy_b3lyp']
y_cc.columns=['energy_cc']
y_b3lyp.head()
```

```python
q_b3lyp.head()
```

```python
q_b3lyp = q_b3lyp.drop([0,2,4,6,8,10,12,14], axis=1)
q_b3lyp.columns=['C1q', 'H1q','H2q','H3q','H4q','C2q','Nq']
q_b3lyp.head()
```

```python
x_b3lyp = x_b3lyp.drop([28], axis=1)
x_cc = x_cc.drop([28], axis=1)
x_cc.head()
print(x_b3lyp.shape)
print(x_cc.shape)
```

```python
xqy_b3lyp = pd.concat([x_b3lyp, q_b3lyp, y_b3lyp], axis=1)
xy_cc = pd.concat([x_cc, y_cc], axis=1)
print(xqy_b3lyp.shape)
print(xy_cc.shape)
```

```python
xy_cc.head()
```

```python
xqy_b3lyp_u = xqy_b3lyp.drop_duplicates(inplace=False)
xy_cc_u = xy_cc.drop_duplicates(inplace=False)
print(xqy_b3lyp_u.shape)
print(xy_cc_u.shape)
```

```python
jcols = list(xqy_b3lyp_u.columns)[0:28]
print jcols
```

```python
# xqy_order = xy_cc_u.merge(right=xqy_b3lyp_u, on=jcols, how="left")
xqy_order = xqy_b3lyp_u.merge(right=xy_cc_u, on=jcols, how="left")
print(xqy_order.shape)
xqy_order.head()
```

```python
ene = xqy_order['energy_cc']
xqy_order.drop(labels=['energy_cc'], axis=1, inplace=True)
xqy_order.insert(35, 'energy_cc', ene)
xqy_order.head()
```

```python
cols = range(0,28,4)
print cols
xqy_new = xqy_order.drop(cols, axis=1)
xqy_new.head()
```

```python
xqy_new.columns=['C1x', 'C1y', 'C1z', 'H1x', 'H1y', 'H1z', 'H2x', 'H2y', 'H2z', 'H3x', 'H3y', 'H3z', 'H4x', 'H4y', 'H4z', 'C2x', 'C2y', 'C2z', 'Nx', 'Ny', 'Nz', 'C1q', 'H1q','H2q','H3q','H4q','C2q','Nq','energy_cc','energy_b3lyp']
print(xqy_new.shape)
```

```python
xqy_new.to_csv('b3lyp_cc_Q_abs.csv')
```

```python
xqy_new.head()
```

```python
import numpy as np
distC = np.sqrt((xqy_new.C1x - xqy_new.C2x)**2 + (xqy_new.C1y - xqy_new.C2y)**2 + (xqy_new.C1z - xqy_new.C2z)**2)
indexMax = distC.idxmax()
maxEneb3lyp = xqy_new["energy_b3lyp"][indexMax]
maxEneCC = xqy_new["energy_cc"][indexMax]
print(indexMax, maxEneb3lyp, maxEneCC)
```

```python
xqy_rel = xqy_new
xqy_rel['energy_b3lyp'] = xqy_new['energy_b3lyp'] - maxEneb3lyp
xqy_rel['energy_cc'] = xqy_new['energy_cc'] - maxEneCC
```

```python
xqy_rel.head()
```

```python
xqy_rel.to_csv('b3lyp_cc_Q_rel.csv')
```

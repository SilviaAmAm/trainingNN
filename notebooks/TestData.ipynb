```python
import pandas as pd
```

```python
x_pbe = x_pbe = pd.read_csv('../dataSets/X_pbe.csv', header=None)
y_pbe = pd.read_csv('../dataSets/Y_pbe.csv', header=None)
q_pbe = pd.read_csv('../dataSets/Q_pbe.csv', header=None)
```

```python
x_b3lyp = x_b3lyp = pd.read_csv('../dataSets/X_b3lyp.csv', header=None)
y_b3lyp = pd.read_csv('../dataSets/Y_b3lyp.csv', header=None)
q_b3lyp = pd.read_csv('../dataSets/Q_b3lyp.csv', header=None)
```

```python
x_pbe.head()
```

```python
x_b3lyp.head()
```

```python
y_pbe.columns=['energy_pbe']
y_b3lyp.columns=['energy_b3lyp']
y_pbe.head()
```

```python
q_pbe.head()
```

```python
q_pbe = q_pbe.drop([0,2,4,6,8,10,12,14], axis=1)
q_pbe.columns=['C1q', 'H1q','H2q','H3q','H4q','C2q','Nq']
q_pbe.head()
```

```python
x_pbe = x_pbe.drop([28], axis=1)
x_b3lyp = x_b3lyp.drop([28], axis=1)
x_b3lyp.head()
```

```python
xqy_pbe = pd.concat([x_pbe, q_pbe, y_pbe], axis=1)
xy_b3lyp = pd.concat([x_b3lyp, y_b3lyp], axis=1)
```

```python
xy_b3lyp.head()
```

```python
xqy_pbe_u = xqy_pbe.drop_duplicates(inplace=False)
xy_b3lyp_u = xy_b3lyp.drop_duplicates(inplace=False)
print(xqy_pbe.shape)
print(xqy_pbe_u.shape)
```

```python
jcols = list(xqy_pbe_u.columns)[0:28]
print jcols
```

```python
xqy_order = xy_b3lyp_u.merge(right=xqy_pbe_u, on=jcols, how="left")
xqy_order.head()
```

```python
ene = xqy_order['energy_b3lyp']
xqy_order.drop(labels=['energy_b3lyp'], axis=1, inplace=True)
xqy_order.insert(35, 'energy_b3lyp', ene)
xqy_order.head()
```

```python
cols = range(0,28,4)
print cols
xqy_new = xqy_order.drop(cols, axis=1)
xqy_new.head()
```

```python
xqy_new.to_csv('pbe_b3lyp_Q_test_abs.csv')
```

```python
xqy_new.columns=['C1x', 'C1y', 'C1z', 'H1x', 'H1y', 'H1z', 'H2x', 'H2y', 'H2z', 'H3x', 'H3y', 'H3z', 'H4x', 'H4y', 'H4z', 'C2x', 'C2y', 'C2z', 'Nx', 'Ny', 'Nz', 'C1q', 'H1q','H2q','H3q','H4q','C2q','Nq','energy_b3lyp','energy_pbe']
```

```python
xqy_new.head()
```

```python
import numpy as np
distC = np.sqrt((xqy_new.C1x - xqy_new.C2x)**2 + (xqy_new.C1y - xqy_new.C2y)**2 + (xqy_new.C1z - xqy_new.C2z)**2)
indexMax = distC.idxmax()
maxEnepbe = xqy_new["energy_pbe"][indexMax]
maxEneb3lyp = xqy_new["energy_b3lyp"][indexMax]
print(indexMax, maxEnepbe, maxEneb3lyp)
```

```python
xqy_rel = xqy_new
xqy_rel['energy_pbe'] = xqy_new['energy_pbe'] - maxEnepbe
xqy_rel['energy_b3lyp'] = xqy_new['energy_b3lyp'] - maxEneb3lyp
```

```python
xqy_rel.head()
```

```python
xqy_rel.to_csv('pbe_b3lyp_Q_test_rel.csv')
```

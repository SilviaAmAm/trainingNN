```python
import pandas as pd
```

```python
x_pbe = x_pbe = pd.read_csv('../dataSets/X_pbe.csv', header=None)
y_pbe = pd.read_csv('../dataSets/Y_pbe.csv', header=None)
q_pbe = pd.read_csv('../dataSets/Q_pbe.csv', header=None)
```

```python
x_b3lyp = x_b3lyp = pd.read_csv('../dataSets/X_b3lyp.csv', header=None)
y_b3lyp = pd.read_csv('../dataSets/Y_b3lyp.csv', header=None)
q_b3lyp = pd.read_csv('../dataSets/Q_b3lyp.csv', header=None)
```

```python
x_pbe.head()
```

```python
x_b3lyp.head()
```

```python
y_pbe.columns=['energy_pbe']
y_b3lyp.columns=['energy_b3lyp']
y_pbe.head()
```

```python
q_pbe.head()
```

```python
q_pbe = q_pbe.drop([0,2,4,6,8,10,12,14], axis=1)
q_pbe.columns=['C1q', 'H1q','H2q','H3q','H4q','C2q','Nq']
q_pbe.head()
```

```python
x_pbe = x_pbe.drop([28], axis=1)
x_b3lyp = x_b3lyp.drop([28], axis=1)
x_pbe.head()
```

```python
xqy_pbe = pd.concat([x_pbe, q_pbe, y_pbe], axis=1)
xy_b3lyp = pd.concat([x_b3lyp, y_b3lyp], axis=1)
```

```python
xqy_pbe.head()
```

```python
xqy_pbe_u = xqy_pbe.drop_duplicates(inplace=False)
xy_b3lyp_u = xy_b3lyp.drop_duplicates(inplace=False)
print(xqy_pbe.shape)
print(xqy_pbe_u.shape)
```

```python
jcols = list(xqy_pbe_u.columns)[0:28]
print jcols
```

```python
xqy = xqy_pbe_u.merge(xy_b3lyp_u, on=jcols, how="inner")
xqy.head()
xqy.shape
```

```python
cols = range(0,28,4)
print cols
```

```python
xqy_new = xqy.drop(cols, axis=1)
xqy_new.head()
```
